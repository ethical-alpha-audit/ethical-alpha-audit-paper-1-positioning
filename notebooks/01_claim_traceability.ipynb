## Context: What Is the Decision-Rule Gap Audit?

Paper 1 conducts a targeted document audit of widely used clinical AI governance instruments. The audit assesses whether each instrument:

1. Specifies an explicit rule for converting multi-domain governance evidence into a binary deployment decision (go/no-go)
2. Defines aggregation logic (e.g., weighted scoring or minimum floors)
3. Specifies any override mechanism and constraints
4. Explicitly assigns final deployment authority

The audit was limited to explicit statements in the instrument text; items were coded as "present" only when decision logic was described in operational terms (e.g., thresholds, pass/fail criteria, or a defined aggregation rule), rather than as high-level recommendations or general lifecycle principles.

The full coding frame is documented in **Multimedia Appendix 4** (`supplementary/Multimedia_Appendix_4_AuditCodingFrame.docx`). The full evidence table with verbatim excerpts and coding decisions is in **Multimedia Appendix 5** (`supplementary/Multimedia_Appendix_5_EvidenceTable.docx`).

## Reading the Table

**Rows 1–6 (existing instruments):** Across all six reviewed instruments, the dominant pattern is:

- **Explicit deployment decision rule:** Not specified in any reviewed instrument as an institutional decision rule
- **Aggregation logic:** Not specified in any reviewed instrument
- **Override pathway:** Not specified in any reviewed instrument  
- **Final deployment authority:** Partial or framework-specific in most; only the EU AI Act and ISO 42001 assign explicit authority (to provider/deployer and management system owner, respectively)

**Row 7 (This Viewpoint):** The proposed gate architecture specifies all four elements — deployment decision rule, aggregation logic (minimum floors), override pathway (constrained), and deployment authority.

### Key Finding

The audit supports a narrow but practically important inference: within the reviewed instruments, the dominant pattern is specification of *what* to evaluate and what lifecycle controls to maintain, with comparatively limited specification of *how* heterogeneous evidence should be operationally combined into a binding deployment decision.


## Interpretive Caveat

> "All characterisations are the author's interpretive assessment."
> — Paper 1 Manuscript

The audit was conducted by a single author using a conservative coding frame (items coded as "present" only when operational decision logic was explicitly stated). This conservatism may overestimate the gap if decision rules exist in sections not audited.
